```r
data <- read.csv('/Users/nicholasbartelo/Documents/AMR Project Final Reviewers/New Analysis Annotated mz/Raw data export of all peaks_no normalization080522_for_maplet.csv', row.names = 2, header =  TRUE)
# Make into CRE and CSE only
data$Group <- ifelse(data$X...Group %like% c("CRE"), "CRE", "CSE")

# Import species info
species_info <- read.csv('/Users/nicholasbartelo/Documents/AMR Project Final Reviewers/New Analysis Annotated mz/Raw_data_with_organism_info.csv')
# Convert rownames of df1 into a column
data$Sample <- rownames(data)
# Merge the dataframes by the 'Sample' column
data_with_organism <- merge(data, species_info[,c("Sample", "Organism")], by = "Sample", all = TRUE)
# Make rownames the sample names
rownames(data_with_organism) <- data_with_organism$Sample

data_with_organism$date_measured <- ifelse(grepl("421", data_with_organism$Sample), "apr21", ifelse(grepl("425", data_with_organism$Sample), "apr25", "apr27"))

# Make list of species to keep
species_to_keep <- c('Klebsiella pneumoniae', 'Enterobacter cloacae', 'Serratia marcescens', 'Proteus mirabilis', 'Klebsiella oxytoca', 'Citrobacter freundii', 'Klebsiella aerogenes', 'Escherichia coli')
# Keep only the desired species
data_with_organism_filtered <- data_with_organism[data_with_organism$Organism %in% species_to_keep, ]

# Create dataframe for metadata for SE object
column_identifiers <- data_with_organism_filtered[, c("Group", "Organism", "date_measured"), drop=FALSE]
# Remove unwanted columns
data_with_organism_filtered$Sample <- NULL
data_with_organism_filtered$Group <- NULL
data_with_organism_filtered$X...Group <- NULL
data_with_organism_filtered$Organism <- NULL
data_with_organism_filtered$date_measured <- NULL

# Transpose dataframe
data_with_organism_filtered = t(data_with_organism_filtered)
# Set all 0 values to missing
data_with_organism_filtered[data_with_organism_filtered == 0] <- NA
# Make dataframe into matrix
data_with_organism_filtered <- as.matrix(data_with_organism_filtered)
```

```r
# Import data
annotations_mz <- read.csv('/Users/nicholasbartelo/Documents/AMR Project Final Reviewers/New Analysis Annotated mz/mapped_data_under_01_tolerance.csv', header =  TRUE)
# Remove X from rownames
rownames(data_with_organism_filtered) <- sub("^X", "", rownames(data_with_organism_filtered))
# Keep only rownames that are in annotated features
data_with_organism_filtered_mz_subset <- data_with_organism_filtered[rownames(data_with_organism_filtered) %in% annotations_mz$X...mz, ]
```

```r
raw_SE <- SummarizedExperiment(assays=list(counts=data_with_organism_filtered_mz_subset), colData = column_identifiers)
processed_SE <- raw_SE %>%
  mt_pre_filter_missingness(feat_max = 0.2) %>% # Filter out features with more than 20% missing
  mt_pre_filter_missingness(samp_max = 0.1) %>%  # Filter out samples with more than 10% missing
  mt_pre_norm_quot() %>% # Quotient Normalization
  mt_pre_trans_log() %>% # Log transform
  mt_pre_trans_scale() %>% # Scale data (z-scores)
  mt_pre_outlier_to_na() %>% # Detect outliers (any values that lie outside the 0.025 and 0.975 regions when assuming a normal distribution)  and set to NA
  mt_pre_impute_knn() # Impute missing values with knn imputation

processed_SE <- processed_SE %>% 
  mt_plots_pca(color = Group)

processed_SE <- processed_SE %>% 
  mt_plots_pca(color = Organism)

processed_SE <- processed_SE %>% 
  mt_plots_pca(color = date_measured)
```

```r
# Make dataframe of processed data
df_processed <- as.data.frame(t(assay(processed_SE)))

# Compute PCA
pca_result <- prcomp(df_processed)

# Calculate the percentage of explained variance for each PC
explained_variance <- (pca_result$sdev^2 / sum(pca_result$sdev^2)) * 100

# Extract the explained variance for PC1 and PC2
pc1_var <- round(explained_variance[1], 2)  # Variance explained by PC1
pc2_var <- round(explained_variance[2], 2)  # Variance explained by PC2

# Prepare metadata for plotting
processed_se_group <- processed_SE@colData$Group
processed_se_organism <- processed_SE@colData$Organism
processed_se_metadata <- data.frame(processed_se_group, processed_se_organism)

# Create a data frame with PCA results and metadata
pca_data <- as.data.frame(pca_result$x)
pca_data <- cbind(pca_data, processed_se_metadata)

# Flip PC1
pca_data$PC1 <- -pca_data$PC1 # To get the data to look like the original PCA plot

# Flip PC2
pca_data$PC2 <- -pca_data$PC2 # To get the data to look like the original PCA plot

final_plot_before_batch_correction_organism_phenotype <- ggplot(pca_data, aes(x = PC1, y = PC2)) +
  geom_point(aes(fill = processed_se_organism, shape = processed_se_group),
             size = 3, stroke = 1, alpha = 0.5) +  # outlines via stroke
  scale_shape_manual(values = c(21, 24), name = "Phenotype") +
  scale_fill_manual(
    values = c("blue", "brown", "gray", "black", "orange", "pink", "yellow", "violet"),
    labels = c(
      expression(italic("Citrobacter freundii")),
      expression(italic("Enterobacter cloacae")),
      expression(italic("Escherichia coli")),
      expression(italic("Klebsiella aerogenes")),
      expression(italic("Klebsiella oxytoca")),
      expression(italic("Klebsiella pneumoniae")),
      expression(italic("Proteus mirabilis")),
      expression(italic("Serratia marcescens"))
    ),
    name = "Organism"
  ) +
  guides(
    fill = guide_legend(override.aes = list(alpha = 0.5, size = 3, stroke = 1, shape = c(21,21,21,21,21,21,21,21))),
    shape = guide_legend(override.aes = list(size = 3, alpha = 0.5, stroke = 1))
  ) +
  theme_minimal() +
 theme(
    panel.border = element_blank(),
    panel.background = element_rect(fill = "white"),
    plot.background = element_rect(fill = "white", colour = NA),
    axis.title.x = element_text(size = 16),
    axis.title.y = element_text(size = 16),
    legend.title = element_text(size = 20),
    legend.text = element_text(size = 20),
    axis.text = element_text(size = 15)
  ) +
  labs(
    x = paste0("Principal Component 1 (", pc1_var, "% variance)"),
    y = paste0("Principal Component 2 (", pc2_var, "% variance)")
  ) +
  coord_fixed()


# Save figure as tiff
ggsave("/Users/nicholasbartelo/Documents/AMR Project Final Reviewers/before_batch_correction_organism_phenotype_pca.tiff", plot = final_plot_before_batch_correction_organism_phenotype, dpi = 300, width = 8, height = 6, units = "in", device = "tiff")
```
